## Hipótese

O nível de desenvolvimento dos países membros da União Europeia é conhecido por ser, geralmente, mais elevados que os países fora do grupo. Os países desse grupo possuem uma maior porcentagem de usuários da internet devido ao nível alto de desenvolvimento? Como a entrada na UE afeta o crescimento de usuários da rede em um país?

A hipótese inicial seria de que, por possuirem uma melhor infraestrutura e desenvolvimento tecnológico, além de serem mais ricos do que grande parte dos outros países do mundo, os membros da União Europeia possuem mais condições de fornecer acesso à internet para uma maior massa de pessoas.

```python
#separação dos países da UE (GDP per capita)
puegdp = gdpc.loc[:,['Austria','Belgium','Bulgaria','Croatia','Cyprus','Czech Republic','Denmark','Estonia','Finland','France','Germany','Greece','Hungary','Ireland','Italy','Latvia','Lithuania','Luxemborg','Malta','Netherlands','Poland','Portugal','Romania','Slovak Republic','Slovenia','Spain','Sweden','United Kingdom']]
puegdp
```

Abaixo iremos separar os dados dos países que ingressaram na UE em 2004:

```python
#separação dos países que ingressaram na UE em 2004 para análise
cyp = pue['Cyprus']
cze = pue['Czech Republic']
est = pue['Estonia']
hng = pue['Hungary']
ltv = pue['Latvia']
lit = pue['Lithuania']
mal = pue['Malta']
pol = pue['Poland']
slk = pue['Slovak Republic']
slv = pue['Slovenia']

#float(cyp[21:]) - float(cyp[8:9])
#cyp[21:]
#cyp[8:9]
```

## Gráfico Inicial

Abaixo plotamos o gráfico que mostra o percentual das populações dos países que ingressaram em 2004 na UE que são usuários de internet ao longo do tempo, para que possamos ver o comportamento geral desse crescimento.

```python
tempo = (1990,1991,1992,1993,1994,1995,1996,1997,1998,1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011)
#pd.plotting.lag_plot(cyp, lag=1, ax=None)
plt.plot(tempo,cyp)
plt.plot(tempo,cze)
plt.plot(tempo,est)
plt.plot(tempo,hng)
plt.plot(tempo,ltv)
plt.plot(tempo,lit)
plt.plot(tempo,mal)
plt.plot(tempo,pol)
plt.plot(tempo,slk)
plt.plot(tempo,slv)
plt.title('Países que entraram na UE em 2004')
plt.xlabel('Ano')
plt.ylabel('Porcentagem de usuários')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

No gráfico acima, podemos ver que as porcentagens tendem a crescer rapidamente a partir de 1995, mas essa curva se torna menos acentuada próximo de 2007. É possível que o gráfico possua uma curva de formato "S", de modo que se estabilize por volta dos 80% ou 90% depois de 2011.

# Alteração Percentual

Com o gráfico acima, conseguimos ver o crescimento da porcentagem de usuários, porém apenas isso não nos ajuda à verificar nossa hipotese. Abaixo, calculamos a alteração porcentual, de modo que que podemos entender como a quantidade de usuários varia com o tempo. Os gráficos foram feitos individualmente para melhor análise.

```python
#cálculo da alteração percentual 
cypp = cyp.pct_change()
czep = cze.pct_change()
estp = est.pct_change()
hngp = hng.pct_change()
ltvp = ltv.pct_change()
litp = lit.pct_change()
malp = mal.pct_change()
polp = pol.pct_change()
slkp = slk.pct_change()
slvp = slv.pct_change()
```

```python
cypp_plot = plt.plot(tempo, cypp)
plt.title('Alteração Percentual do número de usuários em Cyprus')
plt.xlabel('Ano')
plt.ylabel('Alteração Percentual')
plt.show()

czep_plot = plt.plot(tempo, czep)
plt.title('Alteração Percentual do número de usuários na República Tcheca')
plt.xlabel('Ano')
plt.ylabel('Alteração Percentual')
plt.show()

estp_plot = plt.plot(tempo, estp)
plt.title('Alteração Percentual do número de usuários na Estônia')
plt.xlabel('Ano')
plt.ylabel('Alteração Percentual')
plt.show()

hngp_plot = plt.plot(tempo, hngp)
plt.title('Alteração Percentual do número de usuários na Hungria')
plt.xlabel('Ano')
plt.ylabel('Alteração Percentual')
plt.show()

ltvp_plot = plt.plot(tempo, ltvp)
plt.title('Alteração Percentual do número de usuários na Latvia')
plt.xlabel('Ano')
plt.ylabel('Alteração Percentual')
plt.show()

litp_plot = plt.plot(tempo, litp)
plt.title('Alteração Percentual do número de usuários na Lithuania')
plt.xlabel('Ano')
plt.ylabel('Alteração Percentual')
plt.show()

malp_plot = plt.plot(tempo, malp)
plt.title('Alteração Percentual do número de usuários em Malta')
plt.xlabel('Ano')
plt.ylabel('Alteração Percentual')
plt.show()

polp_plot = plt.plot(tempo, polp)
plt.title('Alteração Percentual do número de usuários na Polônia')
plt.xlabel('Ano')
plt.ylabel('Alteração Percentual')
plt.show()

slkp_plot = plt.plot(tempo, slkp)
plt.title('Alteração Percentual do número de usuários na Slovákia')
plt.xlabel('Ano')
plt.ylabel('Alteração Percentual')
plt.show()

slvp_plot = plt.plot(tempo, slvp)
plt.title('Alteração Percentual do número de usuários na Slovênia')
plt.xlabel('Ano')
plt.ylabel('Alteração Percentual')
plt.show()
```

## Análise

Nos gráficos acima, percebemos que há uma grande variação nos primeiros anos, que se estabiliza pelos anos 2002. podemos observar um minúsculo pico na alteração percentual de usuários em quase todos os países em 2004, ano que entraram na UE, mas logo essa alteração volta à cair de modo que fique estável. Esse número pode ou não ser associado à entrada no bloco, portanto, para chegar em resultados mais conclusivos, veremos essa variação em países fora do bloco.
